#Theory Questions:

1. What is statistics, and why is it important?
 - Statistics is the study of collecting, analyzing, interpreting, and presenting data. It's important for making informed decisions based on data.

2. What are the two main types of statistics?
  - Descriptive statistics and inferential statistics.

3. What are descriptive statistics?
  - Descriptive statistics summarize and describe the main features of a dataset (e.g, mean, median, mode).

4. What is inferential statistics?
  - Inferential statistics use a sample to make predictions or generalizations about a population.

5. What is sampling in statistics?
  - Sampling is selecting a subset of individuals from a population to analyze and draw conclusions.

6. What are the different types of sampling methods?
  - Random, stratified, systematic, cluster, and convenience sampling.

7. What is the difference between random and non-random sampling?
  - Random sampling gives every member an equal chance; non-random sampling does not.

8. Define and give examples of qualitative and quantitative data?
  - Qualitative data describes qualities (e.g., color, gender); quantitative data involves numbers (e.g., height, age).

9. What are the different types of data in statistics?
  - Qualitative (nominal, ordinal) and quantitative (interval, ratio).

10. Explain nominal, ordinal, interval, and ratio levels of measurement?

  - Nominal: Categories (e.g., colors)

  - Ordinal: Ordered categories (e.g., rankings)

  - Interval: Numerical, no true zero (e.g., temperature)

  - Ratio: Numerical, true zero (e.g., weight)

11. What is the measure of central tendency?
  - It identifies the center of a dataset, typically via mean, median, or mode.

12. Define mean, median, and mode?

  - Mean: Average

  - Median: Middle value

  - Mode: Most frequent value

13. What is the significance of the measure of central tendency?
  - It helps summarize data with a single representative value.

14. What is variance, and how is it calculated?
 - Variance measures how much data points differ from the mean; calculated as the average of squared deviations.

15. What is standard deviation, and why is it important?
  - It's the square root of variance; it shows how spread out the data is around the mean.

16. Define and explain the term range in statistics?
  - Range is the difference between the highest and lowest values in a dataset.

17. What is the difference between variance and standard deviation?
  - Variance is in squared units; standard deviation is in the same units as the data.

18. What is skewness in a dataset?
  - Skewness indicates the asymmetry of the data distribution.

19. What does it mean if a dataset is positively or negatively skewed?
  - Positive skew: Tail on the right; Negative skew: Tail on the left.

20. Define and explain kurtosis?
  - Kurtosis measures the "tailedness" of the distribution—how peaked or flat it is.

21. What is the purpose of covariance?
  - Covariance shows how two variables vary together (positive or negative relationship).

22. What does correlation measure in statistics?
  - Correlation quantifies the strength and direction of the relationship between two variables.

23. What is the difference between covariance and correlation?
  - Correlation is a standardized form of covariance, bounded between -1 and 1.

24. What are some real-world applications of statistics?
  - Used in healthcare, business, economics, sports, education, and social sciences for data-driven decision-making.

```python
#7. Calculate skewness and kurtosis of a dataset using Python libraries.

from scipy.stats import skew, kurtosis

data = [1, 2, 3, 4, 5, 100]
print("Skewness:", skew(data))
print("Kurtosis:", kurtosis(data))
```

```python
#8.Generate a dataset and demonstrate positive and negative skewness.

from scipy.stats import skewnorm
import seaborn as sns

pos_skew = skewnorm.rvs(10, size=1000)
neg_skew = skewnorm.rvs(-10, size=1000)

sns.histplot(pos_skew, kde=True).set_title("Positive Skew")
plt.show()

sns.histplot(neg_skew, kde=True).set_title("Negative Skew")
plt.show()
```

```python
#9.  Write a Python script to calculate covariance between two datasets.

import numpy as np

x = [1, 2, 3, 4, 5]
y = [5, 4, 6, 7, 8]
cov_matrix = np.cov(x, y)
print("Covariance:", cov_matrix[0, 1])
```

```python
#12. Implement and compare simple random sampling and systematic sampling.
import numpy as np

data = list(range(1, 101))
# Simple random sampling
simple_random = np.random.choice(data, 10, replace=False)

# Systematic sampling
interval = 10
systematic = data[::interval]
print("Random:", simple_random)
print("Systematic:", systematic)
```
