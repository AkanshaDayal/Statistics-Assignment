#Theory Questions:

1. What is a random variable in probability theory?
  - A random variable is a variable whose possible values are numerical outcomes of a random phenomenon.

2. What are the types of random variables?
  - Two types: Discrete (countable values) and Continuous (uncountable, any value in a range).

3. What is the difference between discrete and continuous distributions?
  - Discrete: finite/countable values (e.g. dice rolls), Continuous: infinite values within intervals (e.g. height).

4. What are probability distribution functions (PDF)?
  - A function that gives the probability of a random variable taking a specific value (for discrete) or density (for continuous).

5. How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - CDF gives the probability that a random variable is less than or equal to a value,  PDF gives the likelihood for exact values (or densities).

6. What is a discrete uniform distribution?
  - A distribution where all possible discrete outcomes are equally likely (e.g. rolling a fair die).

7. What are the key properties of a Bernoulli distribution?
  - Two outcomes (0 or 1), with probability p for success and (1 - p) for failure.

8. What is the binomial distribution, and how is it used in probability?
  - Models the number of successes in n independent Bernoulli trials; used in success/failure experiments.

9. What is the Poisson distribution and where is it applied?
  - Models the number of events occurring in a fixed interval of time/space; used in queue theory, call centers, etc.

10. What is a continuous uniform distribution?
  - A distribution where all values in an interval are equally likely (e.g., random number between 0 and 1).

11. What are the characteristics of a normal distribution?
  - Bell-shaped, symmetric around the mean, with mean = median = mode.

12. What is the standard normal distribution, and why is it important?
  - A normal distribution with mean 0 and standard deviation 1; simplifies Z-score calculations and statistical inference.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - It states that the sampling distribution of the sample mean approaches normality as sample size increases, regardless of the population distribution.

14. How does the Central Limit Theorem relate to the normal distribution?
  - CLT explains why the normal distribution is commonly used in statistics, especially for means and proportions.

15. What is the application of Z statistics in hypothesis testing?
  - Z-statistics test hypotheses about population means when the standard deviation is known and the sample size is large.

16. How do you calculate a Z-score, and what does it represent?
  - Z = (X - μ) / σ; it represents how many standard deviations a value is from the mean.

17. What are point estimates and interval estimates in statistics?
  - Point estimate: single value estimate (e.g., sample mean); Interval estimate: range of values (e.g., confidence interval).

18. What is the significance of confidence intervals in statistical analysis?
 - They provide a range of plausible values for a population parameter with a given level of confidence (e.g., 95%).

19. What is the relationship between a Z-score and a confidence interval?
  - Z-scores define the margin of error in a confidence interval (e.g., ±1.96 for 95% confidence).

20. How are Z-scores used to compare different distributions?
  - By standardizing values, Z-scores allow comparison across distributions with different scales or units.

21. What are the assumptions for applying the Central Limit Theorem?

  - Independent observations

  - Identically distributed variables

  - Sample size is sufficiently large (n ≥ 30 is common rule of thumb)

22. What is the concept of expected value in a probability distribution?
  - The long-run average outcome of a random variable, calculated as the sum of all values weighted by their probabilities.

23. How does a probability distribution relate to the expected outcome of a random variable?
  - The expected value is derived from the probability distribution—it represents the theoretical average outcome over many trials.

```python
#1.  Write a Python program to generate a random variable and display its value.
import numpy as np
x = np.random.rand()
print("Random variable:", x)
```

```python
#2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import matplotlib.pyplot as plt
import scipy.stats as stats

x = range(1, 7)
pmf = stats.randint.pmf(x, 1, 7)
plt.stem(x, pmf)
plt.title("PMF - Discrete Uniform (1-6)")
plt.show()
```

```python
#3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
def bernoulli_pdf(x, p):
    return p if x == 1 else 1 - p

print("P(X=1):", bernoulli_pdf(1, 0.6))
```

```python
# 4.  Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram.
data = np.random.binomial(n=10, p=0.5, size=1000)
plt.hist(data, bins=11, density=True)
plt.title("Binomial Distribution")
plt.show()
```

```python
#5.  Create a Poisson distribution and visualize it using Python.
data = np.random.poisson(lam=4, size=1000)
plt.hist(data, bins=range(0, 15), density=True)
plt.title("Poisson Distribution")
plt.show()
```

```python
#6.  Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.
x = np.arange(1, 7)
cdf = stats.randint.cdf(x, 1, 7)
plt.step(x, cdf)
plt.title("CDF - Discrete Uniform")
plt.show()
```

```python
#7. Generate a continuous uniform distribution using NumPy and visualize it.
data = np.random.uniform(0, 1, 1000)
plt.hist(data, bins=20, density=True)
plt.title("Continuous Uniform Distribution")
plt.show()
```

```python
#8.  Simulate data from a normal distribution and plot its histogram.
data = np.random.normal(loc=0, scale=1, size=1000)
plt.hist(data, bins=30, density=True)
plt.title("Normal Distribution")
plt.show()
```

```python
#9.  Write a Python function to calculate Z-scores from a dataset and plot them.
from scipy.stats import zscore
data = np.random.normal(50, 10, 100)
z_scores = zscore(data)
plt.hist(z_scores, bins=20)
plt.title("Z-scores")
plt.show()
```

```python
#10.  Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution.
samples = [np.mean(np.random.exponential(scale=2, size=30)) for _ in range(1000)]
plt.hist(samples, bins=30, density=True)
plt.title("CLT with Exponential Distribution")
plt.show()
```

```python
#11.  Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.
means = [np.mean(np.random.normal(0, 1, 30)) for _ in range(1000)]
plt.hist(means, bins=30, density=True)
plt.title("Sample Means (CLT Verification)")
plt.show()
```

```python
#12. Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1).
x = np.linspace(-4, 4, 100)
pdf = stats.norm.pdf(x, 0, 1)
plt.plot(x, pdf)
plt.title("Standard Normal Distribution")
plt.show()
```

```python
#13. Generate random variables and calculate their corresponding probabilities using the binomial distribution.
x = np.arange(0, 11)
pmf = stats.binom.pmf(x, n=10, p=0.5)
plt.stem(x, pmf)
plt.title("Binomial PMF (n=10, p=0.5)")
plt.show()
```

```python
#14.  Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
x, mean, std = 75, 70, 10
z = (x - mean) / std
print("Z-score:", z)
```

```python
#15.  Implement hypothesis testing using Z-statistics for a sample dataset.
sample = np.random.normal(100, 15, 50)
z = (np.mean(sample) - 100) / (15/np.sqrt(50))
print("Z-statistic:", z)
```

```python
#16.  Create a confidence interval for a dataset using Python and interpret the result.
import scipy.stats as stats
mean = np.mean(sample)
conf_int = stats.norm.interval(0.95, loc=mean, scale=15/np.sqrt(50))
print("95% Confidence Interval:", conf_int)
```

```python
# 17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
data = np.random.normal(100, 10, 100)
mean, std = np.mean(data), np.std(data, ddof=1)
ci = stats.norm.interval(0.95, loc=mean, scale=std/np.sqrt(len(data)))
print("CI for Mean:", ci)
```

```python
#18.  Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.
x = np.linspace(60, 140, 100)
pdf = stats.norm.pdf(x, 100, 15)
plt.plot(x, pdf)
plt.title("Normal PDF")
plt.show()
```

```python
#19.  Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.
x = np.arange(0, 15)
cdf = stats.poisson.cdf(x, mu=4)
plt.step(x, cdf)
plt.title("Poisson CDF")
plt.show()
```

```python
#20.  Simulate a random variable using a continuous uniform distribution and calculate its expected value.
data = np.random.uniform(0, 10, 1000)
print("Expected Value:", np.mean(data))
```

```python
#21.  Write a Python program to compare the standard deviations of two datasets and visualize the difference.
d1 = np.random.normal(50, 5, 100)
d2 = np.random.normal(50, 15, 100)
print("Std Dev D1:", np.std(d1))
print("Std Dev D2:", np.std(d2))
plt.boxplot([d1, d2], labels=['D1', 'D2'])
plt.title("Std Deviation Comparison")
plt.show()
```

```python
#22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.
data = np.random.normal(50, 10, 100)
iqr = np.percentile(data, 75) - np.percentile(data, 25)
print("Range:", np.ptp(data))
print("IQR:", iqr)
```

```python
#23.  Implement Z-score normalization on a dataset and visualize its transformation.
data = np.random.normal(100, 20, 100)
normalized = zscore(data)
plt.hist(normalized, bins=20)
plt.title("Z-score Normalized Data")
plt.show()
```
